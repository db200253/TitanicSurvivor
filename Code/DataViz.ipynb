```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.lines import Line2D
df = pd.read_csv(DATA_PATH)
```

```python
name = ['PassengerId','Survived','Pclass','Sex',
        'Age','SibSp','Parch','Fare']
color = {'0' : 'red', '1' : 'green'}
Dead = df['Survived'] == 0
Alive = df['Survived'] == 1
```

```python
def plotByFeature(name) : 
    legend_elements = [Line2D([0], [0], marker='o', color='red', label='Dead', linestyle="None"),
                    Line2D([0], [0], marker='o', color='green', label='Alive', linestyle="None")]
    fig, ax = plt.subplots()
    ax.legend(handles = legend_elements)
    plt.hist(df[Dead][name],
        facecolor='red',
        bins=50,
        alpha=0.5)
    plt.hist(df[Alive][name],
        facecolor='green',
        bins=50,
        alpha=0.5)
    plt.title('Histogram of ' + name)
    plt.xlabel(name)
    plt.ylabel('Number of death')
    plt.show()
```

```python
for feature in name[2:] :
    plotByFeature(feature)
```

```python
datas = df[name]
sns.pairplot(datas, hue = 'Survived')
```
